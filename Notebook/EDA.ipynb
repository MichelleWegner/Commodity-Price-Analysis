 ### Distribution of Prices

***Silver Price (Green):*** The green line and bars show the distribution of Silver prices, which is the most skewed and has a pronounced peak at a lower price range.

***Conclusion:***

***Oil:*** The bimodal distribution suggests that Oil prices have been volatile, with significant price shifts during the period analyzed.

***Gold:*** Gold prices show less variability compared to Oil, with a strong peak, suggesting a more stable range during the period.

***Silver:*** Silver prices are generally lower than those of Gold and Oil, with a concentrated peak suggesting limited price fluctuation within the analyzed timeframe.

### Price Trends Over Time


in this Plot we Visualizing the price trends of the commodities over time 

```python
#Load necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the merged dataset
file_path = '/Users/mishka/Documents/GitHub/Final-Project/Clean Data/merged_data.csv' 
merged_data = pd.read_csv(file_path)

# Convert 'Date' column to datetime format if not already done
merged_data['Date'] = pd.to_datetime(merged_data['Date'])

# Filter data to only include the years 2008 to 2018
merged_data = merged_data[(merged_data['Date'] >= '2008-01-01') & (merged_data['Date'] <= '2018-12-31')]

# Display the first few rows to verify
print(merged_data.head())

### Price Trends Over Time
# Visualizing the price trends of the commodities over time 
plt.figure(figsize=(10, 6))
plt.plot(merged_data['Date'], merged_data['OIL'], label='Oil Price', color='brown')
plt.plot(merged_data['Date'], merged_data['GLD'], label='Gold Price', color='orange')
plt.plot(merged_data['Date'], merged_data['SLV'], label='Silver Price', color='grey')
plt.title('Price Trends Over Time (2008-2018)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

### Correlation Analysis

We Visulaze the correlation between these commodities so we provide insights into how they move together, which is crucial for portfolio diversification and risk management.

```python
# Calculate correlation matrix
correlation_matrix = merged_data[['OIL', 'GLD', 'SLV']].corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=1, linecolor='black')
plt.title('Correlation Matrix of Commodity Prices')
plt.show()
```

***Correlation Matrix of Commodity Prices***

***Explanation of the Correlation Matrix:***

***Matrix Cells:*** Each cell in the matrix represents the Pearson correlation coefficient between two commodities, ranging from -1 (perfect negative correlation) to 1 (perfect positive correlation).

***Oil and Gold (0.53):*** There is a moderate positive correlation between Oil and Gold prices, suggesting that these commodities tend to move together, though not perfectly.

***Oil and Silver (0.71):*** A stronger positive correlation exists between Oil and Silver prices, indicating that when Oil prices rise, Silver prices are more likely to rise as well.

***Gold and Silver (0.87):*** The strongest correlation is between Gold and Silver, suggesting that these precious metals often move together in the market.

***Conclusion:***

***Gold and Silver:*** The high correlation suggests that these metals are often influenced by similar market factors, making them move together.

***Oil and Silver:*** The moderate-to-strong correlation indicates that Oil price changes are likely to impact Silver, though the relationship is not as strong as between Gold and Silver.

***Oil and Gold:*** The moderate correlation suggests some linkage between Oil and Gold, but other factors also influence their prices independently.
