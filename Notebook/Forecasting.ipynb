# Oil Price Forecasting with Prophet

```python
from prophet import Prophet

# Prepare the data for Prophet - for Oil
df_oil = merged_data[['Date', 'OIL']].rename(columns={'Date': 'ds', 'OIL': 'y'})
model_oil = Prophet()
model_oil.fit(df_oil)
future_oil = model_oil.make_future_dataframe(periods=365)  # Forecasting one year into the future
forecast_oil = model_oil.predict(future_oil)

# Visualize the forecast
fig = model_oil.plot(forecast_oil)
plt.title('Oil Price Forecast')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

***Explanation of the Oil Price Forecast Plot***

***Oil Price Forecast Plot Overview:***
This plot presents the forecasted prices of oil from 2008 to 2018, using historical data and the Prophet model for prediction. The model extrapolates past trends to predict how oil prices might evolve in the future.

***X-Axis (Date):***
The X-axis represents the timeline, starting from 2008 and extending beyond 2018, indicating the dates for which the oil prices were observed and forecasted.

***Y-Axis (Price):***
The Y-axis shows the price of oil, likely in Euro per barrel.

***Black Dots (Historical Data):***
The black dots on the plot signify the actual historical prices of oil during the given period. These data points are used by the Prophet model to identify underlying patterns and forecast future prices.

***Blue Line (Predicted Price):***
The blue line on the plot illustrates the oil prices predicted by the Prophet model. This line follows the trend of historical prices and projects them into the future based on the learned patterns.

***Light Blue Shading (Uncertainty Interval):***
The light blue shading around the predicted price line represents the uncertainty or confidence interval of the predictions. This interval shows the range in which the actual oil prices might lie, with wider intervals indicating greater uncertainty.

***Conclusion:***
The forecast indicates that oil prices have experienced significant fluctuations, with periods of high volatility, particularly during the financial crisis in 2008 and the price drop around 2014-2016. The model suggests a potential recovery in prices post-2018, though with considerable uncertainty, as indicated by the wide confidence intervals.

# ARIMA Model for Time Series Forecasting

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Load the merged dataset
file_path = '/Users/mishka/Documents/GitHub/Final-Project/Clean Data/merged_data.csv'  # Adjust path as necessary
merged_data = pd.read_csv(file_path)

# Prepare the data - Silver prices for ARIMA
silver_prices = merged_data['SLV']

# Fit the ARIMA model (order can be adjusted based on model diagnostics)
arima_model = ARIMA(silver_prices, order=(5, 1, 0))  # (p, d, q)
arima_result = arima_model.fit()

# Forecast the next 365 days
forecast = arima_result.forecast(steps=365)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(silver_prices, label='Actual Silver Prices')
plt.plot(range(len(silver_prices), len(silver_prices) + 365), forecast, label='ARIMA Forecast')
plt.title('Silver Prices Forecast using ARIMA')
plt.xlabel('Date Index')
plt.ylabel('Silver Price')
plt.legend()
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

# Load the merged dataset
file_path = '/Users/mishka/Documents/GitHub/Final-Project/Clean Data/merged_data.csv'
merged_data = pd.read_csv(file_path)

# Prepare the data - Silver prices for LSTM
silver_prices = merged_data['SLV'].values.reshape(-1, 1)

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_silver = scaler.fit_transform(silver_prices)

# Create sequences for LSTM
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

seq_length = 60  # Number of days to consider for each prediction
X, y = create_sequences(scaled_silver, seq_length)

# Split data into training and testing sets
train_size = int(X.shape[0] * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, batch_size=32, epochs=10)

# Make predictions
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(range(seq_length, len(silver_prices)), silver_prices[seq_length:], label='Actual Silver Prices')
plt.plot(range(train_size + seq_length, len(silver_prices)), predictions, label='LSTM Predictions')
plt.title('Silver Prices Forecast using LSTM')
plt.xlabel('Date Index')
plt.ylabel('Silver Price')
plt.legend()
plt.show()
```

***Conclusion :*** The ARIMA model seems to perform best for predicting silver prices, followed by the LSTM model, with the Prophet model trailing in accuracy.
