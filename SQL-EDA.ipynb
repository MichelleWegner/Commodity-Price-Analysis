## Price Changes Over Time Visualization:

```python
import matplotlib.pyplot as plt

# Visualization of Price Changes Over Time
price_changes_data['Date'] = pd.to_datetime(price_changes_data['Date'])

plt.figure(figsize=(10, 6))
plt.plot(price_changes_data['Date'], price_changes_data['OIL'], label='Oil Price')
plt.plot(price_changes_data['Date'], price_changes_data['GLD'], label='Gold Price')
plt.plot(price_changes_data['Date'], price_changes_data['SLV'], label='Silver Price')

plt.title('Price Changes Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

***Explanation of the Price Changes Over Time Chart:***

***This line chart visualizes the price changes of Oil, Gold, and Silver from 2008 to 2018.***

***X-Axis (Date):***
The X-axis represents the timeline from 2008 to 2018, showing the progression of prices over time.

***Y-Axis (Price):***
The Y-axis represents the price of each commodity, presumably in US dollars per unit (e.g., per barrel for oil, per ounce for gold and silver).

***Lines for Each Commodity:***

***Oil Price (blue line):*** The blue line represents the price of oil. You can see significant volatility, particularly around 2008 and 2014, where oil prices experienced sharp declines, reflecting major market events such as the financial crisis and changes in global oil supply.

***Gold Price (orange line):*** The orange line shows the price of gold, which peaks around 2012. This peak corresponds with heightened economic uncertainty, where gold, a traditional safe haven asset, saw a significant rise in price.

***Silver Price (green line):*** The green line tracks the price of silver, which, while lower overall, shows a noticeable spike around 2011. This spike often correlates with similar economic factors influencing gold, though silver is more volatile and industrially driven.

## Summary Statistics Visualization:

```python
import matplotlib.pyplot as plt

# Extract data directly from DataFrame 
min_prices = summary_stats_data[['min_oil_price', 'min_gld_price', 'min_slv_price']].values.flatten()
max_prices = summary_stats_data[['max_oil_price', 'max_gld_price', 'max_slv_price']].values.flatten()
avg_prices = summary_stats_data[['avg_oil_price', 'avg_gld_price', 'avg_slv_price']].values.flatten()

commodities = ['Oil', 'Gold', 'Silver']

plt.figure(figsize=(12, 6))
plt.bar(commodities, max_prices, label='Max Price', alpha=0.7)
plt.bar(commodities, avg_prices, label='Average price', alpha=0.7)
plt.bar(commodities, min_prices, label='Min Price', alpha=0.7)
plt.title('Price ranges of raw materials')
plt.xlabel('Commodities')
plt.ylabel('Price')
plt.legend()
plt.show()
```

***Explanation of Summary Statistics Visualization:***

***X-Axis (Commodities):***
The X-axis represents the three commodities: Oil, Gold, and Silver.

***Y-Axis (Price):***
The Y-axis represents the price in a specific unit (e.g., US dollars per barrel for oil, US dollars per ounce for gold and silver).

***Stacked Bars:***
Each commodity is represented by a stacked bar composed of three segments:

***Min Price (green):*** The bottom segment of each bar shows the minimum price of the commodity during the analyzed period.

***Average Price (orange):*** The middle segment of each bar represents the average price of the commodity during the analyzed period.

***Max Price (blue):*** The top segment of each bar indicates the maximum price of the commodity during the analyzed period.
Interpretation of the Bars:

***Oil:*** Oil prices ranged from approximately 26 (minimum) to 143.95 (maximum), with an average price of about 80.84.

***Gold:*** Gold shows the widest price range, from 70 (minimum) to 184.59 (maximum), with an average price of around 122.67. This indicates that gold experienced the highest price volatility during the analyzed period.

***Silver:*** Silver displays a significantly narrower price range compared to gold and oil, with a minimum of 8.85 and a maximum of 47.26, and an average price of 20.05.

## Correlation matrix Visualization:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# create correlations Dataframe of dataset 
data = {
    'Oil': [1, 0.528654, 0.712958],
    'Gold': [0.528654, 1, 0.867904],
    'Silver': [0.712958, 0.867904, 1]
}
df = pd.DataFrame(data, index=['Oil', 'Gold', 'Silver'])

# Create Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Korrelation zwischen Öl-, Gold- und Silberpreisen')
plt.show()
```

***Explanation of the Correlation Heatmap:***

***The heatmap illustrates the correlations between the prices of three key commodities: Oil, Gold, and Silver.***

***Color Coding:***

***Red:*** Indicates a strong positive correlation (closer to 1).

***Blue:*** Would indicate a strong negative correlation (closer to -1), but there are no negative correlations in this chart.

***Light Red to Orange:*** Represents moderate correlations.

***Values in the Heatmap:***

***Oil and Gold (0.53):*** There is a moderate positive correlation between oil and gold prices. This suggests that when oil prices rise, gold prices tend to rise as well, but the relationship isn't very strong.

***Oil and Silver (0.71):*** The correlation between oil and silver is stronger, indicating that silver prices tend to move more closely with oil prices.

***Gold and Silver (0.87):*** Gold and silver have the strongest correlation, nearly at 0.87. This means that these two precious metals typically move together in the market, reflecting their close relationship in investor sentiment and market conditions.


## Visualization of monthly Price swings:

```python
plt.figure(figsize=(12, 6))

# Plotting the volatility for each commodity over time
plt.plot(monthly_swing_data['month'], monthly_swing_data['oil_volatility'], label='Oil Volatility')
plt.plot(monthly_swing_data['month'], monthly_swing_data['gld_volatility'], label='Gold Volatility')
plt.plot(monthly_swing_data['month'], monthly_swing_data['slv_volatility'], label='Silver Volatility')

plt.title('Monthly Volatility of Commodities')
plt.xlabel('Month')
plt.ylabel('Volatility')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))
plt.tight_layout()
plt.show()
```

***Explanation of the Monthly Volatility of Commodities Chart:***

***This line chart visualizes the monthly volatility of Oil, Gold, and Silver over a specific period.***

***X-Axis (Month):***
The X-axis represents the months in the timeline. Due to the large number of data points, the labels are currently overlapping, making them difficult to read. To improve readability, it would be better to display fewer labels (e.g., every 6 months) or rotate the labels more effectively.

***Y-Axis (Volatility):***
The Y-axis represents the volatility of each commodity. Volatility is a measure of how much the price of each commodity fluctuates over time.

***Lines for Each Commodity:***

***Oil Volatility (blue line):*** The blue line represents the volatility of oil prices. The chart shows significant spikes, indicating periods where oil prices experienced large swings.

***Gold Volatility (orange line):*** The orange line shows the volatility of gold prices. Gold also shows periods of increased volatility, although these are generally less extreme than those of oil.

***Silver Volatility (green line):*** The green line represents the volatility of silver prices. Silver appears to have the lowest volatility of the three commodities, with fewer and smaller spikes.
