```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt 

predictive_indicators_data = pd.read_csv('/Users/mishka/Desktop/FINAL PROJECT/Predictive indicators based on gold price fluctuations.csv')

predictive_indicators_data['Date'] = pd.to_datetime(predictive_indicators_data['Date'])

df_oil = predictive_indicators_data[['Date', 'OIL']].rename(columns={'Date': 'ds', 'OIL': 'y'})
```

```python
model_oil = Prophet()
model_oil.fit(df_oil)

# Prediction for the next 365 days
future_oil = model_oil.make_future_dataframe(periods=365)
forecast_oil = model_oil.predict(future_oil)

# Visualization of Prediction for oil
fig = model_oil.plot(forecast_oil)
plt.title('Oil Price Forecast')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

***Explanation***
This chart illustrates a forecast of oil prices of 10 years based on historical data. 


***X-Axis (Date):*** The X-axis represents the time period over which oil prices were observed, including future predictions for 2018 and beyond.

***Y-Axis (Price):*** The Y-axis shows the oil price, likely in US dollars per barrel.

****Black Dots:**** These represent the actual historical prices for oil.

***Blue Line:*** This line represents the prices predicted by the Prophet model. It follows the historical data and extends into future forecasts.

***Light Blue Shading:*** This is the uncertainty interval (confidence interval) around the forecast. It shows the range within which the true price is likely to fall. The wider the shading, the greater the uncertainty in the forecast.

```python
df_gold = predictive_indicators_data[['Date', 'GLD']].rename(columns={'Date': 'ds', 'GLD': 'y'})

model_gold = Prophet()
model_gold.fit(df_gold)

# Prediction for the next 365 days
future_gold = model_gold.make_future_dataframe(periods=365)
forecast_gold = model_gold.predict(future_gold)

# Visualization of Prediction for Gold
model_gold.plot(forecast_gold)
plt.title('Gold Price Forecast')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

***Explanation of the Gold Price Forecast Plot***

***Gold Price Forecast Plot Overview:***
The plot provides a forecast of gold prices from 2008 to 2018 using historical data. The model used for forecasting is the Prophet model, which is well-suited for time series data and accounts for trends, seasonality, and holidays.

***X-Axis (Date):***
The X-axis represents the timeline, starting from the year 2008 and extending to future predictions beyond 2018. It captures the date on which the gold prices were observed.

***Y-Axis (Price):***
The Y-axis represents the price of gold, presumably in US dollars per ounce.

***Black Dots (Historical Data):***
The black dots represent the actual historical prices of gold observed during the time period. These data points are crucial for training the Prophet model to understand the historical patterns and trends in gold prices.

***Blue Line (Predicted Price):***
The blue line represents the predicted gold prices generated by the Prophet model. This line closely follows the historical data and continues into the future, showing the model's forecast of gold prices for 2018 and beyond.

***Light Blue Shading (Uncertainty Interval):***
The light blue shading around the blue line indicates the uncertainty or confidence interval of the forecast. This shading shows the range within which the actual gold price is likely to fall, given the historical data and model assumptions. A wider shading implies greater uncertainty in the prediction.

***Y-Axis (Price):***

The blue line on the plot represents the silver prices predicted by the Prophet model. This line traces the historical prices and projects future prices based on the learned patterns from the historical data.

***Light Blue Shading (Uncertainty Interval):***
The light blue shading around the predicted price line represents the uncertainty or confidence interval for the predictions. This interval suggests the possible range in which the actual silver prices might fall. Wider intervals indicate higher uncertainty in the predictions.
